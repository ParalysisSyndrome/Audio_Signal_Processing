```python
import numpy as np
import librosa as lb
import librosa.display
import IPython.display as ipd

import matplotlib.pyplot as plt
```

```python
sample = '/content/processed_audio_files/content/processed_audio_files/101_1b1_Al_sc_Meditron_11.wav'

sample_arr, sr = lb.load(sample)
print(sample_arr.shape, sr)
```

```python
ipd.Audio(sample)
```

```python
# Zero-crossing rate with Librosa
FRAME_SIZE = 1024
HOP_LENGTH = 512
zcr = librosa.feature.zero_crossing_rate(sample_arr, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

# Visualise RMSE + waveform
frames = range(len(zcr))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 4))
plt.plot(t, zcr, color="r")
plt.ylim(0, 0.01)
plt.title("RMS")
plt.show()
```
